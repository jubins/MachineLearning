```python
#MLHw3: Problem#3: (A)

tpr = np.array([0, 1/3, 2/3, 1, 1, 1, 1]) #True Positive Rate
fpr = np.array([0, 0, 0, 0, 1/3, 2/3, 1]) #False Positive Rate

plt.plot(fpr,tpr, color='blue', lw=5, label='3A: ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve Graph(Receiver operating characteristic)')
plt.legend(loc='upper right')
plt.show()
print "AUC (Area under the curve) is: 1."
```

```python
#MLHw3: Problem#3: (B)

tpr = np.array([0, 0, 0, 0, 1/3, 2/3, 1]) #True Positive Rate
fpr = np.array([0, 1/3, 2/3, 1, 1, 1, 1]) #False Positive Rate

plt.plot(fpr,tpr, color='blue', lw=5, label='3B: ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve Graph(Receiver operating characteristic)')
plt.legend(loc='upper right')
plt.show()
print "AUC (Area under the curve) is: 0."
```

```python
#MLHw3: Problem#3: (C)

tpr = np.array([0, 1/float(3), 2/float(3), 2/float(3), 1, 1, 1]) #True Positive Rate
fpr = np.array([0, 0, 0, 1/float(3), 1/float(3), 2/float(3), 1]) #False Positive Rate

plt.plot(fpr,tpr, color='blue', lw=5, label='3C: ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve Graph(Receiver operating characteristic)')
plt.legend(loc='upper right')
plt.show()

print "AUC (Area under the curve) is: 8/9."
```
